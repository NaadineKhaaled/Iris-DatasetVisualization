```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.gridspec as gridspec
from sklearn import datasets
from sklearn.decomposition import PCA
import pandas as pd
import seaborn as sn

iris=datasets.load_iris()

iris_classes= iris.target
iris_attributes= iris.data      #rows = values of 4 attributes (150x4)

features= iris.data.T # (4x150)
sepal_length=features[0]
sepal_width=features[1]
petal_length=features[2]
petal_width=features[3]

#-----1.Histogram for each Class-----

print("1- Histogram for each Class: ",end=' ')
colors = ['violet', 'blue','pink']

fig=plt.figure(figsize=(10,10),constrained_layout=True)
spec2 = gridspec.GridSpec(ncols=2, nrows=2, figure=fig)
f1_ax0 = fig.add_subplot(spec2[0, 0])
for label, color in zip(range(len(iris_classes)), colors):  #target iris classes
   # print(label)
    # find the label and plot the corresponding data
    plt.hist(iris_attributes[iris_classes==label, 0],
              label=np.unique(iris.target_names)[label],
              color=color)
plt.title(label="sepal length (cm)")
    
f1_ax1 = fig.add_subplot(spec2[0, 1])
for label, color in zip(range(len(iris_classes)), colors):  #target iris classes
    # find the label and plot the corresponding data
    plt.hist(iris_attributes[iris_classes==label, 1],
              label=np.unique(iris.target_names)[label],
              color=color)
plt.title(label="sepal width (cm)")
    
f1_ax2 = fig.add_subplot(spec2[1, 0])
for label, color in zip(range(len(iris_classes)), colors):  #target iris classes
    # find the label and plot the corresponding data
    plt.hist(iris_attributes[iris_classes==label, 2],
              label=np.unique(iris.target_names)[label],
              color=color)
plt.title(label="petal length (cm)")  

f1_ax3 = fig.add_subplot(spec2[1, 1])
for label, color in zip(range(len(iris_classes)), colors):  #target iris classes
    # find the label and plot the corresponding data
    plt.hist(iris_attributes[iris_classes==label, 3],
              label=np.unique(iris.target_names)[label],
              color=color)
plt.title(label="petal width (cm)")


plt.show()
```

```python
 #----2.Scatter plot 2 attributes together----

fig1=plt.figure(1,figsize=(20,15))
spec3 = gridspec.GridSpec(ncols=2, nrows=2, figure=fig1)
print('\n')
print("2- Scatter plot 2 attributes together: \n ")

colors = ['green', 'orange','black']

f1_ax1 = fig1.add_subplot(spec3[0, 0])
for label, color in zip(range(len(iris_classes)), colors):
  # find the label and plot the corresponding data
  plt.scatter(iris_attributes[iris_classes==label, 0],
              iris_attributes[iris_classes==label, 1],
              label=np.unique(iris.target_names)[label],
              alpha = 1, # transparency
              color=color)
  plt.xlabel(iris.feature_names[0])
  plt.ylabel(iris.feature_names[1])

  plt.legend()

f1_ax2 = fig1.add_subplot(spec3[0, 1])
for label, color in zip(range(len(iris_classes)), colors):
  # find the label and plot the corresponding data
  plt.scatter(iris_attributes[iris_classes==label, 0],
              iris_attributes[iris_classes==label, 2],
              label=np.unique(iris.target_names)[label],
              alpha = 1, # transparency
              color=color)
  plt.xlabel(iris.feature_names[0])
  plt.ylabel(iris.feature_names[2])

  plt.legend()

#----- Another Solution -----------------------------------

f1_ax3 = fig1.add_subplot(spec3[1, 0])
plt.scatter(sepal_length,petal_width,c=iris.target, label=np.unique(iris.target_names)[label])
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[3])


f1_ax4 = fig1.add_subplot(spec3[1, 1])
plt.scatter(sepal_width,petal_length,c=iris.target,label=np.unique(iris.target_names)[label])
plt.xlabel(iris.feature_names[1])
plt.ylabel(iris.feature_names[2])


fig2=plt.figure(2,figsize=(20,10))
spec4 = gridspec.GridSpec(ncols=2, nrows=1, figure=fig2)

f1_ax5 = fig2.add_subplot(spec4[0, 0])
plt.scatter(sepal_width,petal_width,c=iris.target,label=np.unique(iris.target_names)[label])
plt.xlabel(iris.feature_names[1])
plt.ylabel(iris.feature_names[3])
#--------------------------------------------------------------

f1_ax6 = fig2.add_subplot(spec4[0, 1])
for label, color in zip(range(len(iris_classes)), colors):
  # find the label and plot the corresponding data
  plt.scatter(iris_attributes[iris_classes==label,2],
              iris_attributes[iris_classes==label, 3],
              label=np.unique(iris.target_names)[label],
              alpha = 1, # transparency
              color=color)
  plt.xlabel(iris.feature_names[2])
  plt.ylabel(iris.feature_names[3])

  plt.legend()


plt.show()
```

```python
#------ 5.Box Plot to show quartile and median----
num=[0,1,2,3] 
for label,num in zip(range(len(iris_classes)),num):
  fig = plt.figure(figsize =(10, 7))
  ax = fig.add_axes([0, 0, 1, 1])
  data = [iris_attributes[iris.target==0,num],iris_attributes[iris.target==1,num],iris_attributes[iris.target==2,num],iris_attributes[iris.target==3,num]]
  pl =ax.boxplot(data,patch_artist = True)
  plt.xticks([1,2,3],[iris.target_names[0],iris.target_names[1],iris.target_names[2]], rotation ='vertical')
  plt.grid(True)
  for patch, color in zip(pl['boxes'], colors):
    patch.set_facecolor(color)
  
  plt.show()

```

```python
#------3.Scatter 3D plot 3 attributes ------
 

fig3 = plt.figure(1,figsize=(9, 9))

ax = Axes3D(fig3, elev=-150, azim=110)
X_reduced = PCA(n_components=4).fit_transform(iris.data)
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=iris_classes,cmap=plt.cm.Set1,
            edgecolor='k', s=40)
ax.set_title("First three attributes")
ax.set_xlabel(iris.feature_names[0])
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel(iris.feature_names[1])
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel(iris.feature_names[2])
ax.w_zaxis.set_ticklabels([])


fig3 = plt.figure(2,figsize=(9, 9))
ax1 = Axes3D(fig3, elev=-150, azim=110)
ax1.scatter(X_reduced[:, 1], X_reduced[:, 2], X_reduced[:, 3], c=iris_classes,cmap=plt.cm.Set1,
            edgecolor='k', s=40)
ax1.set_title("Second three attributes")
ax1.set_xlabel(iris.feature_names[1])
ax1.w_xaxis.set_ticklabels([])
ax1.set_ylabel(iris.feature_names[2])
ax1.w_yaxis.set_ticklabels([])
ax1.set_zlabel(iris.feature_names[3])
ax1.w_zaxis.set_ticklabels([])

fig3 = plt.figure(3,figsize=(9, 9))
ax2 = Axes3D(fig3, elev=-150, azim=110)
ax2.scatter(X_reduced[:, 0], X_reduced[:, 2], X_reduced[:, 3], c=iris_classes,cmap=plt.cm.Set1,
            edgecolor='k', s=40)
ax2.set_title("Third three attributes")
ax2.set_xlabel(iris.feature_names[0])
ax2.w_xaxis.set_ticklabels([])
ax2.set_ylabel(iris.feature_names[2])
ax2.w_yaxis.set_ticklabels([])
ax2.set_zlabel(iris.feature_names[3])
ax2.w_zaxis.set_ticklabels([])



fig3 = plt.figure(4,figsize=(9, 9))
ax3 = Axes3D(fig3, elev=-150, azim=110)
ax3.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 3], c=iris_classes,cmap=plt.cm.Set1,
            edgecolor='k', s=40)
ax3.set_title("Fourth three attributes")
ax3.set_xlabel(iris.feature_names[0])
ax3.w_xaxis.set_ticklabels([])
ax3.set_ylabel(iris.feature_names[1])
ax3.w_yaxis.set_ticklabels([])
ax3.set_zlabel(iris.feature_names[3])
ax3.w_zaxis.set_ticklabels([])

plt.show()


```
